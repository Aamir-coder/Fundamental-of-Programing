```python
## 12. Simple Calculator.
```

```python
num1 = float(input("Enter First Number:"))
operations = input("Enter Operations:")
num2 = float(input("Enter Second Number:"))
w = (num1 + num2)
x = (num1 - num2)
y = (num1 * num2)
z = (num1 / num2)
if operations == "+":
    print(f': {num1} + {num2} = {w}')
elif operations == "-" :
    print(f': {num1} - {num2} = {x}')
elif operations == "*":
    print(f': {num1} * {num2} = {y}')
elif operations == "/":
    print(f': {num1} / {num2} = {z}')
```
